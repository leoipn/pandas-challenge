```python
#Imports
import pandas as pd
import os
```

```python
file_to_load = os.path.join("Resources", "purchase_data.csv")
```

```python
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data.head()
```

```python
total_count = len(purchase_data["SN"].unique())
```

```python
total_count_df = pd.DataFrame({"Total Players":[total_count]})
total_count_df
```

```python
total_unique_items = len(purchase_data["Item Name"].unique())
avg_price = purchase_data["Price"].mean()
Num_of_purchases = purchase_data["Purchase ID"].count()
total_revenue = purchase_data["Price"].sum()

#pd.options.display.float_format = '${:,.2f}'.format
purchasing_analysis_df = pd.DataFrame({"Number of Unique Items":[total_unique_items],
                                       "Average Price":[avg_price],
                                       "Number of Purchases":[Num_of_purchases],
                                       "Total Revenue":[total_revenue]})

purchasing_analysis_df["Average Price"] = purchasing_analysis_df["Average Price"].map('${:,.2f}'.format)
purchasing_analysis_df["Total Revenue"] = purchasing_analysis_df["Total Revenue"].map('${:,.2f}'.format)

purchasing_analysis_df
```

```python
# Get just the unique players data set
unique_purchase_data = purchase_data.drop_duplicates(subset="SN")

# Group by Genders
grouped_genders = unique_purchase_data.groupby(['Gender'])
```

```python

# Get the total of each gender
total_genders = grouped_genders['SN'].count()
# Get the percentage of each player
percentage_of_players = total_genders*100/total_count

# Show the Gender Demographics Gender
gender_demographics = pd.DataFrame({"Total Count": total_genders,
                                    "Percentage of Players": percentage_of_players})
# Remove the index header
gender_demographics.index.name = None
# Applying float format and % symbol
gender_demographics["Percentage of Players"] =  gender_demographics["Percentage of Players"].map('{:,.2f}%'.format)
# Reordering the rows
gender_demographics = gender_demographics.reindex(["Male","Female","Other / Non-Disclosed"])

# Show the result
gender_demographics
```

```python
# The total info grouped by gender
grouped_total_genders = purchase_data.groupby(['Gender'])

# The total of purchases
purchase_count = grouped_total_genders["SN"].count()

# The average purchase price
avg_purchase_price = grouped_total_genders["Price"].mean()

# The total purchase value
total_purchase_value = grouped_total_genders["Price"].sum()

# The average total purchase per person, this value needs to use the unique users.
avg_total_purchase_per_person = total_purchase_value/total_genders

# The new Data Frame containing the analysis by gender
purchasing_analysis = pd.DataFrame({"Purchase Count": purchase_count,
                                    "Average Purchase Price":avg_purchase_price.map("${:,.2f}".format),
                                    "Total Purchase Value":total_purchase_value.map("${:,.2f}".format),
                                    "Avg Total Purchase per Person":avg_total_purchase_per_person.map("${:,.2f}".format)})
# The info output
purchasing_analysis
```

```python
# Bins for ages
age_bins = [0,9,14,19,24,29,34,39,100]
# Labels for ages
age_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

# Assign to Age group the labels acording to the bin values
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], age_bins, labels=age_labels)
# The duplicated values were dropped
single_purchase_data = purchase_data.drop_duplicates(subset="SN")

# Total count
ages_total_count = single_purchase_data.groupby("Age Ranges")["Age"].count()
# Percentage of players
percent_of_players = (ages_total_count * 100)/total_count

# The age demographics info
age_demographics = pd.DataFrame({"Total Count":ages_total_count,
                                 "Percentage of Players":percent_of_players.map("{:,.2f}%".format)})

# Remove the index header
age_demographics.index.name = None

# The info output
age_demographics
```

```python
grouped_by_age = purchase_data.groupby("Age Ranges")
age_purchase_count = grouped_by_age["Purchase ID"].count()

# The average purchase price
age_avg_purchase_price = grouped_by_age["Price"].mean()

# The total purchase value
age_total_purchase_value = grouped_by_age["Price"].sum()

# The average total purchase per person, this value needs to use the unique users.
age_avg_total_purchase_per_person = age_total_purchase_value/ages_total_count

# The new Data Frame containing the analysis by gender
age_purchasing_analysis = pd.DataFrame({"Purchase Count": age_purchase_count,
                                    "Average Purchase Price":age_avg_purchase_price.map("${:,.2f}".format),
                                    "Total Purchase Value":age_total_purchase_value.map("${:,.2f}".format),
                                    "Avg Total Purchase per Person":age_avg_total_purchase_per_person.map("${:,.2f}".format)})
# The info output
age_purchasing_analysis
```

```python
# The orginal data is grouped by SN
grouped_by_user = purchase_data.groupby('SN')

# Using the grouped data to count the total
sn_total_count = grouped_by_user["SN"].count()

# Using the grouped data to get the average purchase price
sn_avg_purchase_price = grouped_by_user["Price"].mean()

# Using the grouped data to sum the purchase price
sn_total_purchase_price = grouped_by_user["Price"].sum()

# The new Data Frame containing the analysis by SN
user_purchasing_analysis = pd.DataFrame({"Purchase Count": sn_total_count,
                                         "Average Purchase Price":sn_avg_purchase_price,
                                         "Total Purchase Value":sn_total_purchase_price})
# The sorted info
sorted_values = user_purchasing_analysis.sort_values(["Total Purchase Value"], ascending=False)

# The format is applied after the sorting to avoid conflicts with the $ simbol
sorted_values["Average Purchase Price"] = sorted_values["Average Purchase Price"].map('${:,.2f}'.format)
sorted_values["Total Purchase Value"] = sorted_values["Total Purchase Value"].map('${:,.2f}'.format)

# The info output
sorted_values.head()
```

```python
# Data Frame with just 3 columns
items_df = purchase_data[["Item ID", "Item Name", "Price"]]

# Grouped by two columns
group_by_item_id_and_name = items_df.groupby(["Item ID", "Item Name"])

# The items count
item_purchase_count = group_by_item_id_and_name["Price"].count()

# The average item price
item_price = group_by_item_id_and_name["Price"].mean()

# The total item purchase value
total_purchase_value = group_by_item_id_and_name["Price"].sum()

# The new Data Frame containing the analysis by SN
popular_items = pd.DataFrame({"Purchase Count": item_purchase_count,
                              "Item Price":item_price,
                              "Total Purchase Value":total_purchase_value})

# Sorted by purchase count
sorted_popular_items = popular_items.sort_values(["Purchase Count"], ascending=False)

# The format is applied after the sorting to avoid conflicts with the $ simbol
sorted_popular_items["Item Price"] = sorted_popular_items["Item Price"].map('${:,.2f}'.format)
sorted_popular_items["Total Purchase Value"] = sorted_popular_items["Total Purchase Value"].map('${:,.2f}'.format)

# The info output
sorted_popular_items.head()
```

```python
# Items sorted by purchase value
sorted_purchase_value_items = popular_items.sort_values(["Total Purchase Value"], ascending=False)

# The format is applied after the sorting to avoid conflicts with the $ simbol
sorted_purchase_value_items["Item Price"] = sorted_purchase_value_items["Item Price"].map('${:,.2f}'.format)
sorted_purchase_value_items["Total Purchase Value"] = sorted_purchase_value_items["Total Purchase Value"].map('${:,.2f}'.format)

# The info output
sorted_purchase_value_items.head()
```
